So we have equations of the form `m = (2p + 3q)**d and n = (5p + 7q)**d mod N`. Expand these... 
you then get something like 2**-dm - 5**-dn = q**d * (stuff). The point now is to take the gcd of 
the LHS with N. This will yield q. 

```python
## Compute the inverses of 2 and 5 mod N. 
a = pow(2, -d, N)
b = pow(5, -d, N)
# For computing gcd. 
C = a*m - b*n
```

```python
# Result (def in the top code block)
q = gcd(C, N)
p = N // q
```
